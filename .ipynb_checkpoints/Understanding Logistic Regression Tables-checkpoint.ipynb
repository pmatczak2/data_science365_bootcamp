# Basics of logistic regression

# Import the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load the Data

```python
raw_data = pd.read_csv('bank_data.csv', delimiter='\t')
raw_data
```

```python
data = raw_data.copy()
data['y'] = data['y'].map({'yes':1, 'no':0})
```

## Declare the depandent adn independent variables

```python
y = data['y']
x1 = data[['credit', 'march','may','previous']]
```

## Regression 

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
```

## Sumarry

```python
results_log.summary()
```
