## Linear regression model

## Declare the inputs and targets 

```python
targets = data_prepricessed['log_price']
inputs = data_prepricessed.drop(['log_price'], axis=1)
```

## Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

## Create the regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predicitions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title("Eesidual PDF", size=18)
```

```python
reg.score(x_train,y_train)
```

## Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=["Features"])
reg_summary['Wieghts'] = reg.coef_
reg_summary
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predicitions (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf["Target"] - df_pf['Predictions']
```

```python
df_pf["Difference%"] = np.absolute(df_pf["Residual"]/df_pf["Target"]*100)
df_pf
```
